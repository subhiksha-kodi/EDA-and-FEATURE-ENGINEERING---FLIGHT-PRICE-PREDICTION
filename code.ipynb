#FLIGHT PRICE PREDICTION (EDA + FEATURE ENGINEERING)

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline    #make matplotlib plots appear directly below the code cell without needing to explicitly call plt.show().
```

```python
#to open the data in xl, give pip install openpyxl in the terminal (cmd)
data_train=pd.read_excel('Data_Train.xlsx')
data_train
```

```python
test_set=pd.read_excel('Test_set.xlsx')
test_set
```

```python
df=pd.concat([data_train,test_set]).reset_index(drop=True)
df
```

```python
df.info()
```

```python
#Feature Engineering process
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]

#or
#df['Date']=df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
#df['Month']=df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
#df['Year']=df['Date_of_Journey'].apply(lambda x:x.split('/')[2])

df
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head(1)
```

```python
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]

#or
#df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])

df
```

```python
df.isnull().sum()
```

```python
#df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
#df['Arrival_min']=df['Arrival_Time'].str.split(':').str[0]

#or

df['Arrival_hour']=df['Arrival_Time'].apply(lambda x:x.split(':')[0])
df['Arrival_min']=df['Arrival_Time'].apply(lambda x:x.split(':')[1])

df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)

df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(1)
```

```python
df['Dep_hour']=df['Dep_Time'].apply(lambda x:x.split(':')[0])
df['Dep_min']=df['Dep_Time'].apply(lambda x:x.split(':')[1])

df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)

df.head(1)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
df.head(1)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
df
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(3)
```

```python
df['Additional_Info'].unique()
```

```python
df['Duration_hour']=df['Duration'].str.split('h').str[0]
df.head()
```

```python
#converting duration into mins
df['Duration_hour']=df['Duration_hour'].astype(int)  
df.info()
```

```python
df['Duration_hour']=df['Duration_hour']*60
df
```

```python
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.head()
```

```python
#to replace Nan values with 0
df['Duration_min']=df['Duration_min'].replace(np.nan,0)
df.head()
```

```python
df['Duration_min']=df['Duration_min'].astype(int)
df.head()
```

```python
df.info()
```

```python
df['Duration']=df['Duration_hour']+df['Duration_min']
df.head()
```

```python
df.drop(['Duration_hour','Duration_min'],axis=1,inplace=True)
df.head()
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
df['Airline']=labelencoder.fit_transform(df['Airline'])
df['Source']=labelencoder.fit_transform(df['Source'])
df['Destination']=labelencoder.fit_transform(df['Destination'])
df['Additional_Info']=labelencoder.fit_transform(df['Additional_Info'])
```
